```python
df_new.skew()
```

```python
# Now we need to remove the skewness by log function
```

```python
for col in df_new.columns:
    if df_new.skew().loc[col]>0.55:
        df_new[col]=np.log1p(df_new[col])
```

```python
df_new.skew()
```

```python
df_new.head(10)
```

```python
x=df_new.iloc[:,0:9]
x.head(5)
```

```python
y=df_new.iloc[:,-1]
y.head()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
max_r_score=0
for r_state in range(42,105):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=r_state)
    linr=LinearRegression()
    linr.fit(x_train,y_train)
    predlinr=linr.predict(x_test)
    r2score=r2_score(y_test,predlinr)
    if(r2score>max_r_score):
        max_r_score=r2score
        final_r_state=r_state
print("")
print("The maximum value of r2-score corresponding to", final_r_state,'is',max_r_score)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=68)
```

```python
linr=LinearRegression()
```

```python
linr.fit(x_train,y_train)
linr.score(x_train,y_train)
```

```python
predlinr=linr.predict(x_test)
```

```python
print(r2_score(y_test,predlinr))
print(mean_absolute_error(y_test,predlinr))
print('RMSE is: ',np.sqrt(mean_squared_error(y_test,predlinr)))
```

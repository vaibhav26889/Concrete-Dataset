```python
# To treate the Skew values
```

```python
for col in df1.columns:
    if df1.skew().loc[col]>.55:
        df1[col]=np.log1p(df1[col])
```

```python
df1.skew()
```

```python
df1.head(10)
```

```python
x=df1.iloc[:,0:-1]
x.head()
```

```python
y=df1.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
max_r_score=0
for r_state in range(40,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=r_state)
    linr=LinearRegression()
    linr.fit(x_train,y_train)
    predlinr=linr.predict(x_test)
    r2scr=r2_score(y_test,predlinr)
    if (r2scr>max_r_score):
        max_r_score=r2scr
        final_r_state=r_state

print("")
print('The max r2-Score corresponding to the',final_r_state,'is',max_r_score)
```
